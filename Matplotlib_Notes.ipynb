# Matplotlib example workflow

```python
# 0. Import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(3, 3)) # width and height

# 3. Plot data
ax.plot(x, y)

# 4. Customize plot
ax.set(title="Simple plot", xlabel='X-axis', ylabel='Y-axis')

# 5. Save & show figure
fig.savefig("../content/sample_data/imgs/sample-plot.png")
```

# Making figures with NumPy arrays
* Line plot
* Scatter plot
* Bar plot
* Histogram
* Subplots

```python
import numpy as np
```

```python
# create data
x = np.linspace(0, 10, 100)
```

## *line plot*

```python
# plot data and create a line plot
fig, ax = plt.subplots(figsize=(3, 3))
ax.plot(x, x**2)
plt.show()
```

## *Scatter plot*


```python
# use same data to scatter
fig, ax = plt.subplots(figsize=(3, 3))
ax.scatter(x, np.exp(x))
plt.show()
```

```python
# anoter scatter
fig, ax = plt.subplots(figsize=(3, 3))
ax.scatter(x, np.sin(x))
plt.show()
```

## *Bar plot*
* Bar (vertical)
* Barh (horizontal)


```python
# make bar plot from dictionary
car_prices = {"BMW": 100,
              "Honda": 90,
              "Toyota": 120}
fig, ax = plt.subplots(figsize=(3, 3))
ax.set(title="Cars Prices", xlabel="names",ylabel="prices")
ax.bar(car_prices.keys(), car_prices.values())
plt.show()
```

```python
# barh
fig, ax = plt.subplots(figsize=(3, 3))
ax.barh(list(car_prices.keys()), list(car_prices.values()))
plt.show()
```

## *Histogram plot*

```python
x = np.random.randn(1000)
fig, ax = plt.subplots(figsize=(3, 3))
ax.hist(x)
plt.show()
```

## *Subplot (2 options)*

```python
# Option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                            ncols=2,
                                            figsize=(5,5))
# plot to each different axis
ax1.plot(x, x/2)
ax2.scatter(np.random.random(10), np.random.random(10))
ax3.bar(car_prices.keys(), car_prices.values())
ax4.hist(np.random.randn(1000))
plt.show()
```

```python
# Option 2
fig, ax = plt.subplots(nrows=2,
                       ncols=2,
                       figsize=(5, 5))
# Plot to each different index
ax[0, 0].plot(x, x/2)
ax[0, 1].scatter(np.random.random(10), np.random.random(10))
ax[1, 0].bar(car_prices.keys(), car_prices.values())
ax[1, 1].hist(np.random.randn(1000))
plt.show()
```

## pyplot vs. matplotlib OO method (Object Oriented)
* when plotting somthing *quickly*, use (pyplot) method.
* when plotting somthing *more advanced*, use (OO) method.

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head()
```

## *Pyplot method*

```python
over_50.plot(x='age',
             y='chol',
             c='target',
             kind='scatter',
             figsize=(4, 3));
```

## *OO method*

```python
# OO method mixed with pyplot method
fig, ax = plt.subplots(figsize=(4, 3)) # ax is the where we added data
over_50.plot(x='age',
             y='chol',
             c='target',
             kind='scatter',
             ax=ax);
# Set limts to the X-axis
# ax.set_xlim([45, 100])
```

```python
# Create OO method from scratch
fix, ax = plt.subplots(figsize=(4, 3))

# plot data
scatt = ax.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"]);

# customize the plot
ax.set(title="Heart disease and chol levels",
       xlabel="Age",
       ylabel="Chol");

# Add legend
ax.legend(*scatt.legend_elements(), title="Target");

# Add a horizontal line
ax.axhline(over_50["chol"].mean(),
           linestyle='--');
```

```python
heart_disease.head()
```

```python
# Subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2,
                               ncols=1,
                               figsize=(5, 6));

# Add title to the entire figure
fig.suptitle("Heart disease analysis", fontsize=16, fontweight="bold");

# Add data to ax0
scatt = ax0.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"],
                    cmap="winter"); # change the scheme
# Customize ax0
ax0.set(title="Heart disease and chol levels",
        xlabel="Age",
        ylabel="Chol");

# Change the X-axim lim
ax0.set_xlim([50, 80])
# Add legend to ax0
ax0.legend(*scatt.legend_elements(), title="Target")
# Add a meanline
ax0.axhline(y=over_50["chol"].mean(),
            linestyle='--');


# Add data to ax1
scatt = ax1.scatter(x=over_50["age"],
                    y=over_50["thalach"],
                    c=over_50["target"],
                    cmap="winter");
# customize ax1
ax1.set(title="Heart disease and thalach(max heart rate)",
        xlabel="Age",
        ylabel="Thalach");
# Change the X-axim lim
ax1.set_xlim([50, 80])
ax1.set_ylim([60, 200])
# Add legend to ax1
ax1.legend(*scatt.legend_elements(), title="Target")
# Add a meanline
ax1.axhline(y=over_50["thalach"].mean(),
            linestyle='--');
```
